```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
colour = np.random.rand(50)
```

```python
Scatter Plot is nothing but plotting scatters on a graph using data.
This creates a visualization, where we can't really visuzlise by looking just on data points.
Scatters can be a Dot(.), X, V,etc.
This is how we can generate scatter plot
```

```python
plt.scatter(x,y , c= colour)
plt.xlabel("This is my X label")
plt.ylabel("This is my Y label")
plt.title("This is my Title")
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
A subplot is a function of matplotlib library. It is used to create multiple plot within a single figure.=
```

```python
fig,axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0,0].plot(x, y1,label="Line 1")
axes[0,1].plot(x, y2,label="Line 2")
axes[1,0].plot(x, y3,label="Line 3")
axes[1,1].plot(x, y4,label="Line 4")

#For X & Y Axis Label
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('Y-axis')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')

#For Title
axes[0,0].set_title("Line 1")
axes[0,1].set_title("Line 2")
axes[1,0].set_title("Line 3")
axes[1,1].set_title("Line 4")
```
